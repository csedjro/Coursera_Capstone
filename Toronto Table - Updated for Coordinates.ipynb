```python
import pandas as pd

url='https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'

df=pd.read_html(url, header=0)[0]

df.head()
```

```python
df.drop(df[df['Borough']=='Not assigned'].index, inplace = True)
df.head()
```

```python
df
```

```python
## Neigbourhood are already grouped by Borough
```

```python
df.groupby(['Postal Code','Borough']).agg({'Neighbourhood':lambda x: list(x)})
```

```python
df['Neighbourhood'].replace('Not assigned',df['Borough'])
```

```python
df_Final = df.groupby(['Postal Code','Borough']).agg({'Neighbourhood':lambda x: list(x)})
```

```python
df_Final.head()
```

```python
## verifying dimension of the file
df_Final.shape
```

```python
## importing the csv file with coordinates
df_coordinates = pd.read_csv(r'C:\Users\csedj\Downloads\Geospatial_Coordinates.csv')
df_coordinates.head()
```

```python
## merging the two files based on postal code
df_Merged = pd.merge(df_Final, df_coordinates, on='Postal Code')
```
